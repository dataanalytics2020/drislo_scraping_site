```python
import pandas as pd
from datetime import datetime

# カスタム関数を定義して日付をソート可能な形式に変換
def convert_to_sortable_date(date_str):
    month, day = map(int, date_str.split('(')[0].split('/'))
    current_year = datetime.now().year
    date = datetime(current_year, month, day)
    if date > datetime.now():
        date = date.replace(year=current_year - 1)
    return date

# groupby_hall_daily_dfのイベント日をソート
groupby_hall_daily_df['ソート用日付'] = groupby_hall_daily_df['イベント日'].apply(convert_to_sortable_date)
groupby_hall_daily_df = groupby_hall_daily_df.sort_values('ソート用日付',ascending=False)
```
