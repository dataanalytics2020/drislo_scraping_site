```python
extract_card_elems_df = card_elems_df[card_elems_df['prefecture'].isin(['東京都', '神奈川県', '千葉県', '埼玉県'])]
extract_card_elems_df.reset_index(drop=True, inplace=True)
extract_card_elems_df
```

```python
groupby_hall_daily_df= pd.DataFrame(index=[], columns=[])
groupby_hall_status_df = pd.DataFrame(index=[], columns=[])
import traceback
for i ,row in extract_card_elems_df.iterrows():
    print(row['link'])
    browser.get(row['link'])
    browser.implicitly_wait(10)
    html = browser.page_source.encode('utf-8')
    soup = BeautifulSoup(html, 'html.parser')
    click_link_elms = soup.find_all('div', attrs={ 'class': 'subpages__accordion-arrow' })
    hall_name = row['hall_name']
    prefix_hall_name = row['prefix_hall_name']
    print(hall_name)
    click_link_cards = browser.find_elements(By.XPATH, '//*[@class="card"]')
    all_hall_daily_df = pd.DataFrame(index=[], columns=[])
    all_hall_status_df = pd.DataFrame(index=[], columns=[])
    
    for i, click_link_card_elem in enumerate(click_link_cards):
        try:
            if  i > 5:
                break
            #要素をクリック
            #print(click_link_card_elem)
            card_table_html = click_link_card_elem.get_attribute('innerHTML')
            card_table_html = BeautifulSoup(card_table_html, 'html.parser')
            
            card_header = click_link_card_elem.find_element(By.XPATH, '//*[@class="card-header"]')
            event_date = card_table_html.find('p',attrs={'class': 'subpages__accordion-date'}).text
            #print(event_date)

            syuzai_list = []
            for img_src in card_table_html.find_all('img', attrs={ 'class': 'subpages__accordion-img'}):
                #print(img_src.get('src'))
                if 'dc31.png' in img_src.get('src') :
                    syuzai_list.append('スマスロ・6.5号機')
                elif 'fcc6.png' in img_src.get('src') :
                    syuzai_list.append('三台並び')
                elif '5ee1.png' in img_src.get('src') :
                    syuzai_list.append('ジャグラー・ハナハナ')
                elif 'ef94.png' in img_src.get('src') :
                    syuzai_list.append('全台')
                elif 'dc35.png' in img_src.get('src') :
                    syuzai_list.append('スマスロ・4.0号機')
                else:
                    syuzai_list.append('その他')
            #print(syuzai_list)
            report_button = click_link_card_elem.find_element(By.XPATH, '//*[@class="subpages__accordion-arrow"]')
            report_button.click()
            time.sleep(1)
            browser.execute_script('window.scrollTo(0, document.body.scrollHeight);')
            #htmlを取得

            dfs = pd.read_html(str(card_table_html))

            status_df  = dfs[0].T
            #0行目をカラム名にする
            status_df.columns = status_df.iloc[0]
            #0行目を削除
            try:
                status_df.drop(0, axis=0,inplace=True)
            except:
                status_df.reset_index( inplace=True)
                status_df.drop(0, axis=0,inplace=True)
            status_df['イベント日'] = event_date
            status_df['公約'] = str(syuzai_list)
            status_df.reset_index(drop=True, inplace=True)
            status_df

            #h2タグのテキストを取得
            card_table_html = click_link_card_elem.get_attribute('innerHTML')
            h2_soup = BeautifulSoup(card_table_html, 'html.parser')
            h2_text_list = h2_soup.find_all('h2')
            reversed_h2_text_list =list(reversed(h2_text_list))


            #0行目をカラム名にする
            total_status_df = status_df.T
            total_status_df.columns = total_status_df.iloc[0]
            #0行目を削除
            try:
                total_status_df.drop(0, axis=0,inplace=True)
            except:
                total_status_df.reset_index( inplace=True)
            total_status_df.reset_index(drop=True, inplace=True)
            total_status_df

            all_status_df = status_df.join([total_status_df])
            all_hall_status_df = pd.concat([all_hall_status_df, all_status_df],axis=0)

            daily_status_df = pd.DataFrame(index=[], columns=[])
            h2_num = 0
            
            new_reversed_h2_text_list = []
            for h2_text in reversed_h2_text_list:
                if 'style' in h2_text.attrs:
                    new_reversed_h2_text_list.append(h2_text.text)
            new_reversed_h2_text_list
            for _df in list(reversed(dfs))[:-1]:
                #display(_df)
                if '台番号' in _df.columns:
                    _df['イベント日'] = event_date
                    _df['公約'] = str(syuzai_list)
                    _df['エリア名'] = hall_name
                    _df['ホール名'] = prefix_hall_name
                    try:
                        _df['内容'] = new_reversed_h2_text_list[h2_num]
                    except:
                        _df['内容'] = new_reversed_h2_text_list[-1]
                    h2_num += 1
                    daily_status_df = pd.concat([daily_status_df, _df])
                    
            daily_status_df['ゲーム数'] = daily_status_df['ゲーム数'].map(lambda x: x.replace('G', '').replace(',', '').replace('±', ''))
            daily_status_df['差枚'] = daily_status_df['差枚'].map(lambda x: x.replace('枚', '').replace(',', '').replace('±', ''))
            daily_status_df['ゲーム数'] = daily_status_df['ゲーム数'].astype(int)
            daily_status_df['差枚'] = daily_status_df['差枚'].astype(int)
            daily_status_df['台番号'] = daily_status_df['台番号'].astype(int)
            daily_status_df
            all_hall_daily_df = pd.concat([all_hall_daily_df, daily_status_df],axis=0)
        except Exception as e:
            error_message = traceback.format_exc()  
            print(hall_name,event_date,error_message)
            #break
            continue
    groupby_hall_status_df = pd.concat([groupby_hall_status_df, all_hall_status_df],axis=0)
    groupby_hall_daily_df = pd.concat([groupby_hall_daily_df, all_hall_daily_df],axis=0)
        
    #break
```
